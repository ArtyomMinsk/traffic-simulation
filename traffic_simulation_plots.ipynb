# Road Rage
###### Our program simulates 30 cars traveling on a continuous 1km stretch of road for one minute. Our task was to reccomend the speed limit for this road based on our findings. 

```python
position, speeds = main()
```

### Car Positions
###### The following graph shows the position of each car at every second that the simulation runs. It is evident that there is a traffic jam around second 30 that seems to effect traffic through the end of the simulation; however, by second 60, the spacing between the cars begins to even out. 

```python
y = position
x = list(range(len(position)))
v = [-1, 61, -1, 1001]
for x, y in zip(x, y):
    plt.scatter([x] * len(y), y)

plt.axis('tight')
plt.tight_layout()
plt.ylabel('Position on Road (meters)')
plt.xlabel('Time in Simulation (seconds)')
plt.title('Car Positions Throughout Simulation')
sns.despine()
plt.show()
```

### Car Speeds
###### We decided to show a graphical representation of the speed of each car as the minute progressed.  The dots represent multiple cars while the squares represent one car.  One of the conclusions we could determine from this representation is that it is very rare for all the cars to be traveling at the max speed (33m/s) at once.  As the seconds progress, the range of speeds increases as well.  

```python
y = speeds
x = list(range(len(position)))
for x, y in zip(x, y):
    plt.scatter([x] * len(y), y)

plt.axis('tight')
plt.tight_layout()
plt.ylabel('Speed (meters/second)')
plt.xlabel('Time in Simulation (seconds)')
plt.title('Car Speeds Throughout Simulation')
sns.despine()
plt.show()

```

## Recommended Speed Limit
###### Through our simulation, we were able to conclude that 32 m/s would be the best speed limit for this road. We determined the speed limit by finding the average speed and adding one standard deviation to it.  Such a speed limit was set due to the fact that it is a straight stretch of road that does not allow passing. 

```python
avg_speed_list = [sum(item) / len(item) for item in speeds] 
mean_speed = np.mean(avg_speed_list)
subtraction = [np.array(item) - mean_speed for item in avg_speed_list]
squared = [np.array(item) ** 2 for item in subtraction]
standard_deviation = math.sqrt(sum(squared) / len(squared))
speed_limit = int(mean_speed + standard_deviation)
print('Speed limit: ', speed_limit)
```
